```python
def drop_file(data,pregunta,valor):
    data = data[data[pregunta] != valor]
    return data
```

```python
#Estas variables la escala en la cual están medidas es inversa,es decir, la mejor valoración es el número 
#mas bajo, por lo que voy a múltiplicarlas por -1 para dar la vuelta a este fenómeno
inver_valores=("P1","P2","P3","P801","P802","P803","P804")
```

```python
def inversion(data,columna):
    data[columna] = data[columna]*-1
    return data
```

```python
for columna in inver_valores:
    encuesta = inversion(encuesta,columna)
```
